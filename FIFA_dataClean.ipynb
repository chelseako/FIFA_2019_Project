# Remove money symbols and convert to full value

```python
def moneyConverter(Value):
    if isinstance(Value,str):
        out = Value.replace('€', '')
        if 'M' in out:
            out = float(out.replace('M', ''))*1000000
        elif 'K' in Value:
            out = float(out.replace('K', ''))*1000
        return float(out)
```

```python
fifa['Value'] = fifa['Value'].apply(lambda x: moneyConverter(x))
fifa['Wage'] = fifa['Wage'].apply(lambda x: moneyConverter(x))
fifa['Release Clause'] = fifa['Release Clause'].apply(lambda x: moneyConverter(x))
```

# Clean height

```python
def cleanHeight(x):
    'Returns height in total inches'
    
    if isinstance(x,str):
        l=x.split("'")
        i=int(l[0])
        j=int(l[1])
        s=(i*12)+j
    return(s)

fifa['Height'] = fifa['Height'].apply(cleanHeight)
```

```python
fifa['Height']
```

# Clean weight

```python
def cleanWeight(x):
    'Removes lbs from value'
    
    if isinstance(x,str):
        return(x.replace('lbs', ''))
    return(x)

fifa['Weight'] = fifa['Weight'].apply(cleanWeight).astype('float')
```

```python
fifa['Weight']
```

# Bin position

```python
def binPosition(x):
    'Bins positions into forward, midfielder, defender'
    
    forward = ['LW', 'LF', 'ST', 'CF', 'RF', 'RW', 'LS', 'RS']
    midfielder = ['CAM', 'LM', 'CM', 'CDM', 'RM', 'LAM', 'LCM', 'LDM', 'RAM', 'RCM', 'RDM']
    defender = ['CB', 'LB', 'RB', 'LWB', 'RWB', 'LCB', 'RCB']
    
    
    if isinstance(x,str):
        if x in forward:
            return('Forward')
        elif x in midfielder:
            return('Midfielder')
        elif x in defender:
            return('defender')
        else:
            return(-1)

fifa['Position'] = fifa['Position'].apply(binPosition)
fifa['Position']
```

# Clean skill rating variables

```python
def skillConverter(val):
    'Adds skill rating to create one number'
    if type(val) == str:
        s1 = val[0:2]
        s2 = val[-1]
        val = int(s1) + int(s2)
        return val
    else:
        return val

skillColumns = ['LS', 'ST', 'RS', 'LW', 'LF', 'CF', 'RF', 'RW', 'LAM', 'CAM',
       'RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM', 'LWB', 'LDM', 'CDM', 'RDM',
       'RWB', 'LB', 'LCB', 'CB', 'RCB', 'RB']

for col in skillColumns:
    fifa[col] = fifa[col].apply(skillConverter)
```
